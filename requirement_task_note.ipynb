# Streamlit Multipage App Implementation Notes

## Overview

We've restructured our application to use Streamlit's native multipage app functionality. This allows for better organization of code and a more intuitive user experience with standard navigation.

## Structure

The application is organized as follows:

```
DBMS_InsuranceManagement/
├── Home.py                       # Main entry point
├── pages/                        # Streamlit pages directory
│   ├── 1_Dashboard.py            # Dashboard page
│   ├── 2_Customer_Management.py  # Customer management
│   ├── 3_Insurance_Types.py      # Insurance types
│   ├── 4_Contract_Management.py  # Contract management
│   ├── 5_Claims_Assessments.py   # Claims handling
│   ├── 6_Payouts_Management.py   # Payouts management
│   └── 7_Reports.py              # Reports and analytics
├── models/                       # Data models (optional)
├── database/                     # Database connectivity
│   └── db_connector.py           # Database connector with caching
└── ...
```

## Key Implementation Points

1. **Page Naming Convention**: Pages are prefixed with numbers to control their order in the sidebar.

2. **Data Caching**: Database queries use `@st.cache_data` to improve performance:
   ```python
   @st.cache_data(ttl=300)  # Cache data for 5 minutes
   def get_data():
       # Query the database
       return data
   ```

3. **Database Connection Management**: All database connections are properly closed after use.

4. **Error Handling**: Database errors are handled gracefully with appropriate user feedback.

5. **Consistent UI**: All pages follow a consistent UI pattern with similar layouts and component usage.

## Navigation

Streamlit automatically creates a sidebar with links to each page based on the files in the `pages` directory. The order is determined by the numeric prefixes in the filenames.

1. **Query Caching**: Database queries are cached for 5 minutes to reduce database load
2. **Connection Pooling**: Database connections are created when needed and closed properly
3. **Clear Cache on Updates**: Cache is cleared automatically after write operations

# Code Organization Strategy: UI-Logic Separation

To maintain a clean and maintainable codebase, we've adopted a specific code organization strategy:

## 1. Directory Structure

```
DBMS_InsuranceManagement/
├── Home.py                     # Main entry point
├── pages/                      # UI components only
│   ├── 1_Dashboard.py          # Dashboard UI
│   ├── 2_Customer_Management.py
│   └── ...
├── models/                     # Business logic and database queries
│   ├── dashboard.py            # Dashboard data functions
│   ├── customer.py             # Customer-related queries
│   └── ...
└── database/                   # Database connectivity
    └── db_connector.py         # Database connection handling
```

## 2. Responsibilities:

### Pages Directory (UI Layer)
Files in the `pages` directory are responsible for:
- Defining the page layout
- Rendering UI components
- Handling user interactions
- Calling model functions to get data
- Formatting and displaying data

### Models Directory (Business Logic Layer)
Files in the `models` directory are responsible for:
- Containing all SQL queries
- Data processing and transformation
- Business rules implementation
- Database interaction
- Caching strategies via `@st.cache_data` decorators

## 3. Benefits of This Approach:

- **Separation of Concerns**: UI code and business logic are kept separate
- **Code Reusability**: Database queries can be reused across different pages
- **Maintainability**: Easier to modify SQL queries without affecting the UI
- **Testing**: Simpler to test business logic without UI dependencies
